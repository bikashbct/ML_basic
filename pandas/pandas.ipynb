```python
import pandas as pd
```

```python
df = pd.read_csv("credit_card_fraud_10k.csv", encoding="utf-8" )
```

```python
data = df.head(200)
df= pd.DataFrame(data)
print(f"Shape,{df.shape}")
print(f"Columns,{df.columns}")
```

```python
#handling with missing data
print(df.isnull().sum())
```

```python
df['merchant_category'].fillna(df['merchant_category'].mode()[0], inplace=True)
df.head(21)
df.to_csv("credit_card_fraud_10k.csv", index=False)
df.sort_values(by='transaction_id', ascending=True, inplace=True)
df.to_csv("credit_card_fraud_10k.csv", index=False)
df.sort_values(by = 'amount', ascending=True, inplace=True)
```

```python
#data aggregation summary statistics
mean_amount = df['amount'].mean()
print(f"Mean transaction amount: {mean_amount}")
max_amount = df['amount'].max()
print(f"Max transaction amount: {max_amount}")
min_amount = df['amount'].min()
print(f"Min transaction amount: {min_amount}")
fraud_count = df['is_fraud'].sum()
print(f"Total fraudulent transactions: {fraud_count}")
non_fraud_count = df.shape[0] - fraud_count
print(f"Total non-fraudulent transactions: {non_fraud_count}")
```

```python
#grouping and pivoting data
grouped = df.groupby(['merchant_category'])['is_fraud'].sum()
print(grouped)
```

```python
#merging and joining datasets
dataframe1 = pd.DataFrame(df)
df = pd.read_csv("fraud_grocery.csv", encoding="utf-8" )
dataframe2 = pd.DataFrame(df)
merged_df = pd.merge(dataframe1, dataframe2, on='transaction_id', how='inner', suffixes=('_left', '_right'))
print(merged_df.head(10))
merged_df.to_csv("merged_fraud_data.csv", index=False)
```

```python
# joining\
left_joined_df = pd.merge(dataframe1, dataframe2, on='transaction_id', how='left', suffixes=('_left', '_right'))
left_joined_df.to_csv("left_joined_fraud_data.csv", index=False)
```

```python
#concatenating data
concatenated_df = pd.concat([dataframe1, dataframe2], axis=0, ignore_index=True)
concatenated_df.to_csv("concatenated_fraud_data.csv", index=False)
```
