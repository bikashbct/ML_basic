```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.linspace(0,5,11)
y = x **3
```

```python
plt.plot(x, y )
plt.title("Plot of x vs y")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

```python
#subplots
plt.subplot(2,2,1)
plt.plot(x,y,'-b')
plt.subplot(2,2,2)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(x,x**2)
plt.subplot(2,2,4)
plt.plot(y,x)
```

```python
#object oriented method
fig = plt.figure(figsize=(5,5), dpi=100)
axis_1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis_1.plot(x,y)
axis_1.plot(y,x)
```

```python
plt.scatter(x,y)
```

```python
plt.hist(y)
```

```python
plt.boxplot(y)
```

```python
fig = plt.figure()
fig, axes = plt.subplots(nrows=1, ncols=2)
for ax in axes:
    ax.plot(x,y)
plt.savefig('myplot.png')    
```
