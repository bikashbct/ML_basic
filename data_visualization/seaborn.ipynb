```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Load an example dataset
df= sns.load_dataset("tips")
```

```python
df['size'].unique()
```

```python
#displot - distribution plot
plt.subplot(1,2,1)
data = sns.histplot(df['total_bill'],bins= 20, kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins= 20, kde=True)
```

```python
#joint plot 
sns.jointplot(x='total_bill', y='tip', data=df, kind='scatter')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='hex')
```

```python
#pair plot
sns.pairplot(df,hue="size", palette="rainbow")
```

```python
#rug plot
sns.rugplot(df['total_bill'])
```

```python
#categorial plot
#boxplot
sns.boxplot(x='day', y='total_bill', data=df, hue='sex', palette='rainbow')
```

```python
sns.countplot(x=df['sex'], hue=df['smoker'])
              
```

```python
sns.barplot(x='day', y='total_bill', data=df, hue='sex', estimator=np.sum)
```

```python
sns.violinplot(x='day', y='total_bill', data=df, hue='sex')
```

```python
sns.stripplot(x='day', y='total_bill', data=df, hue='sex', jitter=True, dodge=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data=df, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=df)
sns.swarmplot(x='day', y='total_bill', data=df, color='pink')
```

```python
#matrix plots 
flights= sns.load_dataset("flights")
fp= flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
#cluster map
sns.clustermap(tips)
```

```python
#regression plot
sns.lmplot(x='total_bill', y='tip', data=df, hue='smoker', palette='rainbow', markers=['o','v'], scatter_kws={'s':100})
```
